ÐžÑÑ‚Ð°ÐµÑ‚ÑÑ Ñ€Ð°Ð·Ð¾Ð±Ñ€Ð°Ñ‚ÑŒÑÑ ÑÐ¾ ÑÑ‚Ð¾Ð»Ð±Ñ†Ð¾Ð¼ abilities

```python
import re
df["abilities"].value_counts()

def transform(s):
    s = s.lower()
    s = re.sub('[^a-z, ]','', s)
    return s.split(', ')

df["abilities"] = df["abilities"].apply(transform)
```

```python
s = set()
for abilities in df["abilities"]:
    for j in abilities:
        s.add(j)
```

```python
abilities = list(s) 
X = np.zeros((len(df["abilities"]),len(abilities)))
for j in range(len(df["abilities"])): # Ð¿Ñ€Ð¾Ð¹Ð´ÐµÐ¼ÑÑ Ð¿Ð¾ ÐºÐ°Ð¶Ð´Ð¾Ð¼Ñƒ Ð¿Ð¾ÐºÐµÐ¼Ð¾Ð½Ñƒ    
    for ability in df["abilities"].iloc[j]: # Ð¿Ñ€Ð¾Ð¹Ð´ÐµÐ¼ÑÑ Ð¿Ð¾ ÐºÐ°Ð¶Ð´Ð¾Ð¹ ÑÐ¿Ð¾ÑÐ¾Ð±Ð½Ð¾ÑÑ‚Ð¸ Ð¿Ð¾ÐºÐµÐ¼Ð¾Ð½Ð°
        i = abilities.index(ability) # Ð¸Ð½Ð´ÐµÐºÑ Ð²Ñ…Ð¾Ð¶Ð´ÐµÐ½Ð¸Ñ ÑÑ‚Ð¾Ð¹ ÑÐ¿Ð¾ÑÐ¾Ð±Ð½Ð¾ÑÑ‚Ð¸
        X[j][i] = 1 # Ð´Ð»Ñ j-Ð³Ð¾ Ð¿Ð¾ÐºÐµÐ¼Ð¾Ð½Ð° i-Ð°Ñ ÑÐ¿Ð¾ÑÐ¾Ð±Ð½Ð¾ÑÑ‚ÑŒ

    
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split(df, Y, test_size = 0.2)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```
