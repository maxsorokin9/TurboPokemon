Столбцы с пропусками: height_m, percentage_male, type2, weight_kg

```python
# в столбец height_m на месте пропусков добавим среднее по столбцу
size = sum(df["height_m"].isna() == False) # количество строк без пропусков
summ = np.array(df["height_m"])[df["height_m"].isna() == False].sum() # сумма элементов без пропусков
mean = summ / size
df["height_m"] = df["height_m"].fillna(mean)
```

```python
# нам сказано, что покемон бесполый, если у него пропуск в столбце, поэтому заполним пропуск 0.5 - сколько 
# самцов, столько и самок
df["percentage_male"] = df["percentage_male"].fillna(0.5)
```

```python
#пропуск в столбце type2 заполним просто пустой строкой
df["type2"] = df["type2"].fillna("")
```

```python
# в столбец weight_kg на месте пропусков добавим среднее по столбцу
size = sum(df["weight_kg"].isna() == False) # количество строк без пропусков
summ = np.array(df["weight_kg"])[df["weight_kg"].isna() == False].sum() # сумма элементов без пропусков
mean = summ / size
df["weight_kg"] = df["weight_kg"].fillna(mean)
```

Избавились от пропусков в данных

Категориальные признаки у нас classification, type1, type2, abilities

Для столбцов classification, type1, type2 применим лэйблы, каждому классу или типу просто будет ставить число

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df["classfication"])
df["classfication"] = le.transform(df["classfication"])
le.fit(df["type1"])
df["type1"] = le.transform(df["type1"])
le.fit(df["type2"])
df["type2"] = le.transform(df["type2"])
```

Остается разобраться со столбцом abilities

```python
import re
df["abilities"].value_counts()

def transform(s):
    s = s.lower()
    s = re.sub('[^a-z, ]','', s)
    return s.split(', ')

df["abilities"] = df["abilities"].apply(transform)
```

```python
s = set()
for abilities in df["abilities"]:
    for j in abilities:
        s.add(j)
```

```python
abilities = list(s) 
X = np.zeros((len(df["abilities"]),len(abilities)))
for j in range(len(df["abilities"])): # пройдемся по каждому покемону    
    for ability in df["abilities"].iloc[j]: # пройдемся по каждой способности покемона
        i = abilities.index(ability) # индекс вхождения этой способности
        X[j][i] = 1 # для j-го покемона i-ая способность

    
```

```python
df1 = pd.DataFrame(X, columns = abilities)
df1
```

```python
df.drop(["abilities"], axis = 1, inplace = True)
```

```python
df = df.join(df1)
```

## Обучим модель

```python
Y = df["is_legendary"]
Y.value_counts() # мало легендарных, и много обычных
```

```python
df.drop(["is_legendary"], inplace = True, axis = 1)
```

```python
df.drop(["name"], inplace = True, axis = 1)
```

```python
df["capture_rate"].iloc[773] = 142
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split(df, Y, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, Y_train)
y_pred = tree.predict(x_test)
y_pred_proba = tree.predict_proba(x_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```
